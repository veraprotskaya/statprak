## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:


````python
import matplotlib.pyplot as plt
import numpy as np

def beauty_derivative_plot(xtang):
    x = np.linspace(-6.5, 6.5, 1000)
    plt.axis([-6.5, 6.5, -4, 4])

    def f(x):
        return 2 * np.sin(x) - (np.cos(2 * x))

    def slope(x):
        return 2*np.cos(x)+2*np.sin(2*x)

    y1=f(xtang)
    #y=m*(x-x1)+y1
    def tangentline(x,xtang,y1):
        return slope(xtang)*(x-xtang)+y1

    xrange=np.linspace(xtang-1,xtang+1,10)

    plt.plot(x, f(x),"b",label="f(x)=2sinx-cos2x")
    plt.plot(x,slope(x),"r--",label="f'(x)=2cosx+2sin2x")
    plt.scatter(xtang,y1,color='black',s=50)
    plt.plot(xrange,tangentline(xrange,xtang,y1),linewidth=2,label=f"Tangent at x={float(xtang)}",color="g")
    plt.annotate(f"f ' ({xtang}) = {round(slope(xtang),2)}",xy=(xtang,y1),xytext=(xtang-2,y1-2),color="black",fontsize=8,arrowprops={"facecolor": "black","edgecolor": "black", "shrink": 0.05, "width": 0.5, "headwidth": 4, "alpha": 0.7,"connectionstyle": "arc3,rad=0.1"})

    plt.axhline(y=0,xmin=-6.5,xmax=6.5,linewidth=1,color="gray")
    plt.axvline(x=0, ymin=-4, ymax=4,linewidth=1,color="gray")

    plt.plot([xtang, xtang], [0, y1],color="purple",linestyle="--",alpha=0.5)
    plt.plot([0, xtang],[y1, y1],color="purple",linestyle="--",alpha=0.5)
    plt.title('Function, Derivative, and Tangent Line')
    plt.legend(loc="upper right",fontsize='small')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()


beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(1.7)


````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
import matplotlib.pyplot as plt
import numpy as np

arr = np.array([[1900, 30000, 4000, 48300],
                [1901, 47200, 6100, 48200],
                [1902, 70200, 9800, 41500],
                [1903, 77400, 35200, 38200],
                [1904, 36300, 59400, 40600],
                [1905, 20600, 41700, 39800],
                [1906, 18100, 19000, 38600],
                [1907, 21400, 13000, 42300],
                [1908, 22000, 8300, 44500],
                [1909, 25400, 9100, 42100],
                [1910, 27100, 7400, 46000],
                [1911, 40300, 8000, 46800],
                [1912, 57000, 12300, 43800],
                [1913, 76600, 19500, 40900],
                [1914, 52300, 45700, 39400],
                [1915, 19500, 51100, 39000],
                [1916, 11200, 29700, 36700],
                [1917, 7600, 15800, 41800],
                [1918, 14600, 9700, 43300],
                [1919, 16200, 10100, 41300],
                [1920, 24700, 8600, 47300]])
evr = {'year': arr[:, 0], 'hare': arr[:, 1], 'lynx': arr[:, 2], 'carrot': arr[:, 3]}

years = evr['year']
hares = evr['hare']

mask = (years >= 1917) & (years < 1920)
hares_recent = hares[mask]

hare_estimated_1920 = np.mean(hares_recent)

index_1920 = np.where(years == 1920)[0][0]
hare_true_1920 = hares[index_1920]

error_ratio = abs(hare_estimated_1920 - hare_true_1920) / hare_true_1920

print(f"Estimated hare population in 1920 (mean method): {hare_estimated_1920:.2f}")
print(f"True hare population in 1920: {hare_true_1920:.2f}")
print(f"Error ratio: {error_ratio:.4f}")
````
